```python
!pip install sqlalchemy==1.3.9
!pip install ipython-sql
!pip install ipython-sql prettytable

%load_ext sql

import csv, sqlite3
import prettytable
prettytable.DEFAULT = 'DEFAULT'

con = sqlite3.connect("my_data1.db")
cur = con.cursor()

!pip install -q pandas

%sql sqlite:///my_data1.db

import pandas as pd
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
df.to_sql("SPACEXTBL", con, if_exists='replace', index=False,method="multi")

#DROP THE TABLE IF EXISTS

%sql DROP TABLE IF EXISTS SPACEXTABLE;

%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null

#Task 1: Display the names of the unique launch sites in the space mission

%sql SELECT * FROM SPACEXTABLE;

%sql SELECT DISTINCT Launch_Site FROM SPACEXTABLE;

#task 2: Display 5 records where launch sites begin with the string 'CCA'
## This is my first try: print(head, "Launch_Site" starts with 'CCA')

# Filter rows where 'Launch_Site' starts with 'CCA'
filtered_data = df[df['Launch_Site'].str.startswith('CCA')]

# Display the first 5 records
print(filtered_data.head())

## Task 3:Display the total payload mass carried by boosters launched by NASA (CRS)

#to sum the payload mass
sum_payload_mass = df['PAYLOAD_MASS__KG_'].sum()
print(sum_payload_mass)

#ti filter by NASA(CRS)
filtered_data = df[df['Customer'] == 'NASA (CRS)']

#to sum the payload mass
sum_payload_mass = filtered_data['PAYLOAD_MASS__KG_'].sum()
print(sum_payload_mass)

#Task 4: Display average payload mass carried by booster version F9 v1.1
average_payload = df['PAYLOAD_MASS__KG_'].mean()
filtered_data = df[df['Booster_Version'] == 'F9 v1.1']

# Calculate the average payload mass for F9 v1.1 boosters
average_payload_f9v1_1 = filtered_data['PAYLOAD_MASS__KG_'].mean()

print(average_payload_f9v1_1)

#task 5:List the date when the first succesful landing outcome in ground pad was acheived.
##Hint:Use min function
min_Date = df['Date'].min()
filtered_data = df[df['Landing_Outcome'] == 'Success (ground pad)']
min_Date = filtered_data['Date'].min()
print(min_Date)

#Task 6
#List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000
payload=df[(df['Landing_Outcome'] == 'Success (drone ship)') & (df['PAYLOAD_MASS__KG_'] > 4000) & (df['PAYLOAD_MASS__KG_'] < 6000)]
print(payload)

#Task 7: List the total number of successful and failure mission outcomes¶

## Task7.1
filtered_data = df[df['Mission_Outcome'] == 'Success']
print(len(filtered_data))
print(df['Mission_Outcome'].eq('Success').sum())

## Task7.2
filtered_data = df[df['Mission_Outcome'] == 'Failure']
print(len(filtered_data))
print(df['Mission_Outcome'].eq('Failure').sum())

#wow, you can read my mind!!! So cool!

#Task 8: List the names of the booster_versions which have carried the maximum payload mass. Use a subquery

max_Payload = df['PAYLOAD_MASS__KG_'].max()
booster_version = df.loc[df['PAYLOAD_MASS__KG_'] == max_Payload, 'Booster_Version'].values[0]

print(f"Max Payload: {max_Payload} kg, Booster_Version: {booster_version}")

##Task 9
#List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.
#Note: SQLLite does not support monthnames. So you need to use substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.
filtered_month = df[(df['Date'] >= '2015-01-01') & (df['Date'] <= '2015-12-31')]
filtered_month = filtered_month[filtered_month['Landing_Outcome'] == 'Failure (drone ship)']
filtered_month = filtered_month[['Date', 'Booster_Version', 'Launch_Site']]
print(filtered_month)

##Task 10
#Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.
filtered = df[(df['Date'] >= '2010-06-04') & (df['Date'] <= '2017-03-20')]
filtered = filtered['Landing_Outcome'].value_counts().sort_values(ascending=False)
print(filtered)


```
